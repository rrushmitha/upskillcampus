```python
import matplotlib.pyplot as plt

traffic = df.groupby('Junction')['Vehicles'].mean()

traffic.plot(kind='bar')

plt.title('Average Vehicles by Junction')

plt.xlabel('Junction')

plt.ylabel('Average Vehicles')

plt.show()
```
