```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/datafile (1).csv')

df.head()
```

```python
print(df.columns.tolist())
```

```python
df.columns = df.columns.str.strip()
```

```python
df.groupby('Crop')['Yield (Quintal/ Hectare) '].mean().plot(kind='bar')
```

```python
df.columns = df.columns.str.strip()
print(df.columns.tolist())
```

```python
import matplotlib.pyplot as plt

df.groupby('Crop')['Yield (Quintal/ Hectare)'].mean().plot(
    kind='bar',
    figsize=(10,5)
)

plt.title("Average Yield by Crop")
plt.xlabel("Crop")
plt.ylabel("Yield (Quintal/Hectare)")
plt.xticks(rotation=45)
plt.show()
```

```python
df.groupby('State')['Yield (Quintal/ Hectare)'].mean().plot(
    kind='bar',
    figsize=(12,5)
)

plt.title("Average Yield by State")
plt.xlabel("State")
plt.ylabel("Yield")
plt.xticks(rotation=90)
plt.show()
```

```python
print("Number of Unique Crops:", df['Crop'].nunique())
print(df['Crop'].unique())
```

```python
print("Number of States:", df['State'].nunique())
print(df['State'].unique())
```

```python
yield_col = [col for col in df.columns if "Yield" in col][0]

highest = df.loc[df[yield_col].idxmax()]

print(highest[['Crop', 'State', yield_col]])
```

```python
cost_col = 'Cost of Cultivation (`/Hectare) C2'

df.groupby('Crop')[cost_col].mean().plot(
    kind='bar',
    figsize=(10,5)
)

plt.title("Average Cultivation Cost by Crop")
plt.xlabel("Crop")
plt.ylabel("Cost")
plt.xticks(rotation=45)
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])

print(numeric_df.corr())
```

```python
import matplotlib.pyplot as plt

corr = numeric_df.corr()

plt.figure(figsize=(6,5))
plt.imshow(corr, cmap='coolwarm', interpolation='nearest')
plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Heatmap")
plt.show()
```
